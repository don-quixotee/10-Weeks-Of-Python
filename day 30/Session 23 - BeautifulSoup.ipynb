```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.flipkart.com/mobiles/pr?sid=tyy,4io&marketplace=FLIPKART")
print(response)
```

```python
200 : Success 

Client side : 
401 : Unauthorised
403 : Permission denied 
404 Page Not Found

Server Side:
500 : Internal Server error
```

```python
print(response.status_code)
print(response.content)
```

```python
soup = BeautifulSoup(response.content,"html.parser")
print(soup)
```

```python
Top - Btm
find("tag_name",attrs = {"id":123,"class":"abc"},text ="ABC") : fetch first_occ of that tag in given page
find_all("tag_name") : fetch all the occ of given tag 
  
Btm - Top 
find_parent('tag_name')
find_parents('tag_name')

Sideways
find_sibling('tag')
find_siblings('tag')
```

```python
name = soup.find("div",attrs = {"class":"_3wU53n"})
price = soup.find("div",attrs= {"class":"_2rQ-NK"})

print(price.text)
```

```python
card = soup.find("div",attrs = {"class":"_1UoZlX"})

name = card.find("div",attrs = {"class":"_3wU53n"})
price = card.find("div",attrs= {"class":"_2rQ-NK"})
rating = card.find("div",attrs = {"class":"hGSR34"})
print("{} {} {}".format(name.text,price.text,rating.text))
```

```python
all_names = soup.find_all("div",attrs = {"class":"_3wU53n"})
for name in all_names:
    print(name.text)
```

```python
cards = soup.find_all("div",attrs = {"class":"_1UoZlX"})

for card in cards:
    name = card.find("div",attrs = {"class":"_3wU53n"})
    price = card.find("div",attrs= {"class":"_2rQ-NK"})
    rating = card.find("div",attrs = {"class":"hGSR34"})
    print("{} {} {}".format(name.text,price.text,rating.text))
```

```python
link = soup.find("a",text = "Next").get("href")
home_page_url = "https://www.flipkart.com"
next_page_link = home_page_url+link
# print(next_page_link)

response2 = requests.get(next_page_link)
soup2 = BeautifulSoup(response2.content,"html.parser")

cards = soup2.find_all("div",attrs = {"class":"_1UoZlX"})

for card in cards:
    name = card.find("div",attrs = {"class":"_3wU53n"})
    price = card.find("div",attrs= {"class":"_2rQ-NK"})
    rating = card.find("div",attrs = {"class":"hGSR34"})
    print("{} {} {}".format(name.text,price.text,rating.text))
```
