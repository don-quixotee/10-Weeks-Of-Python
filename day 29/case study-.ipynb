```python
class Property:
    def __init__(self, square_feets = '', beds ='', baths ='', **kwargs):
        super().__init__(**kwargs)
        self.area = square_feets
        self.numOfbeds = beds
        self.numOfbaths = baths
        
    def display(self):
        print("-"*50)
        print()
        print("PROPERTY DEATAILS")
        print("_"*50)
        print()
        print("Area : {}".format(self.area))
        print("Bedrooms : {}".format(self.numOfbeds))
        print("Bathrooms : {}".format(self.numOfbaths))
        print("_"*50)
    def prompt_init():
        
        """ here we are gonna create a static method to take the input deatails and pass 
            it as an argument
        """
        return dict(square_feets = input("Enter the square feet : "),
                   beds = input("Enter number of bedrooms : "),
                   baths = input("enter the numbers of bathrooms : "))
    prompt_init = staticmethod(prompt_init)
    
    
    
```
